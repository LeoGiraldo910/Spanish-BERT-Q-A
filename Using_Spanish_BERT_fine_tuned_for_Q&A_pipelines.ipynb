````
from transformers import *
````

## 1. Create the pipeline

````
# By now we do not use fast tokenizer so we create the tokenizer with the object {"use_fast": False}
nlp = pipeline(
    'question-answering',
    model='mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es',
    tokenizer=(
        'mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es',
        {"use_fast": False}
    )
)
````

## 2. Set the context

````
context = 'Había una vez un niño llamado Mateo, un curioso pequeño explorador de ocho años. Siempre estaba ansioso por descubrir nuevos lugares y aprender cosas emocionantes. Un día, encontró un mapa antiguo en el ático de su abuela, que parecía llevar a un lugar misterioso en el bosque.Decidido a explorar este lugar secreto, Mateo se preparó para la aventura. Tomó una mochila, llenó una cantimplora con agua y se puso su sombrero de explorador. Siguió el mapa a través del espeso bosque, atravesando densas ramas y arbustos. Finalmente, llegó a un claro donde descubrió una cueva escondida.Mateo entró en la cueva con cautela y, para su asombro, encontró un mundo subterráneo lleno de estalactitas y estalagmitas brillantes. Caminó más adentro y se encontró con un arroyo subterráneo que fluía entre rocas relucientes. Decidió seguir el arroyo y descubrió un tesoro de cristales resplandecientes en una sala subterránea.Pasó horas explorando este mundo subterráneo y recogiendo algunos de los cristales para llevarlos de regreso a casa como recuerdo. Cuando finalmente decidió regresar a la superficie, se dio cuenta de que se había perdido en la cueva. Sin embargo, Mateo recordó lo que su abuela siempre le decía: "Enfrenta tus miedos y busca soluciones"Con valentía, Mateo usó sus conocimientos de orientación para encontrar su camino de regreso a la entrada de la cueva. Emergió del bosque con una historia emocionante que contar a su familia y una mochila llena de cristales brillantes.'

````

## 3. Make questions

````
nlp(
    {
        'question': '¿Cómo se llama el niño de la historia? ',
        'context': context
    }
)
````

````
nlp(
    {
        'question': '¿Dónde encontró Mateo el mapa antiguo?',
        'context': context

    }
)
````

````
nlp(
    {
        'question': '¿Qué descubrió Mateo en la cueva subterránea?',
        'context': context

    }
)
````
